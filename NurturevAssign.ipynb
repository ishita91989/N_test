## A
Which account has the biggest decline in Sales team over the last 6 months ?
Should we consider the absolute or relative decline ? Why ?

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sample_data/pdl_file.csv')
df.head()
```

```python
df.info()
```

```python
from ast import literal_eval
from pandas.tseries.offsets import DateOffset
```

```python
def parse_employee_counts(employee_count_str):
    try:
        return literal_eval(employee_count_str)
    except:
        return {}

df['employee_count_by_month_by_role_parsed'] = df['employee_count_by_month_by_role'].apply(parse_employee_counts)
#parsing in each row
```

```python
def extract_sales_data(employee_data):
    sales_data = {}
    for month, roles in employee_data.items():
        if 'sales' in roles:
            sales_data[month] = roles['sales']
    return sales_data

df['sales_data'] = df['employee_count_by_month_by_role_parsed'].apply(extract_sales_data)
```

```python
df[['domain_name', 'sales_data']].head()
#'extracted sales data'
```

```python
def calculate_sales_decline(sales_data):
    sales_df = pd.DataFrame(list(sales_data.items()), columns=['month', 'count'])
    sales_df['month'] = pd.to_datetime(sales_df['month'], format='%Y-%m')
    sales_df = sales_df.set_index('month').sort_index()

    six_months_ago = pd.to_datetime('today') - DateOffset(months=6)
    recent_sales = sales_df[sales_df.index > six_months_ago]

    if not recent_sales.empty:
        start_count = recent_sales.iloc[0]['count']
        end_count = recent_sales.iloc[-1]['count']

        #absolute and relative decline
        absolute_decline = end_count - start_count
        if start_count != 0:
            relative_decline = (absolute_decline / start_count) * 100
        else:
            relative_decline = float('nan')

        return start_count, end_count, absolute_decline, relative_decline
    return None, None, None, None
```

```python
df['start_sales'], df['end_sales'], df['absolute_decline'], df['relative_decline'] = zip(*df['sales_data'].apply(calculate_sales_decline))

print(df[['domain_name', 'start_sales', 'end_sales', 'absolute_decline', 'relative_decline']].head())
```

```python
largest_relative_decline = df.loc[df['relative_decline'].idxmin()]
```

```python
print(f"Company with the largest relative sales decline: {largest_relative_decline['domain_name']}")
print(f"Relative Decline: {largest_relative_decline['relative_decline']}%")
print(f"Start Sales: {largest_relative_decline['start_sales']}, End Sales: {largest_relative_decline['end_sales']}")
```

We should consider Relative Decline. Relative decline is generally better analysis as it works good when companies are of different size.
It is relatively good for doing broader comparisons. It provides a percentage change that is more comparable across different organizations.

## B
Which companies have a recent executive arrival/departures in engineering teams ?
Create a dataframe with the columns: website x title x date joined x movement_type (arrival/departure)



```python
from ast import literal_eval
```

```python
def parse_exec_data(exec_str):
    try:
        return literal_eval(exec_str)
    except:
        return []
```

```python
df['recent_exec_hires_parsed'] = df['recent_exec_hires'].apply(parse_exec_data)
df['recent_exec_departures_parsed'] = df['recent_exec_departures'].apply(parse_exec_data)
```

```python
def extract_movements(exec_data, movement_type, website):
    movements = []
    for entry in exec_data:
        job_title = entry['job_title'].lower()
        if 'engineering' in job_title or 'technology' in job_title:
            movements.append({
                'website': website,
                'title': entry['job_title'],
                'date_joined': entry.get('joined_date') if movement_type == 'arrival' else entry.get('departed_date'),
                'movement_type': movement_type
            })
    return movements
```

```python
engineering_movements = []
for idx, row in df.iterrows():
    engineering_movements += extract_movements(row['recent_exec_hires_parsed'], 'arrival', row['website'])
    engineering_movements += extract_movements(row['recent_exec_departures_parsed'], 'departure', row['website'])
```

```python
engineering_movements_df = pd.DataFrame(engineering_movements)
print(engineering_movements_df)
```

```python
engineering_movements_df.to_csv('output_table.csv', index=False)
pd.read_csv('output_table.csv')
```
