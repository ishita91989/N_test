## B
Which companies have a recent executive arrival/departures in engineering teams ?
Create a dataframe with the columns: website x title x date joined x movement_type (arrival/departure)



```python
from ast import literal_eval
```

```python
def parse_exec_data(exec_str):
    try:
        return literal_eval(exec_str)
    except:
        return []
```

```python
df['recent_exec_hires_parsed'] = df['recent_exec_hires'].apply(parse_exec_data)
df['recent_exec_departures_parsed'] = df['recent_exec_departures'].apply(parse_exec_data)
```

```python
def extract_movements(exec_data, movement_type, website):
    movements = []
    for entry in exec_data:
        job_title = entry['job_title'].lower()
        if 'engineering' in job_title or 'technology' in job_title:
            movements.append({
                'website': website,
                'title': entry['job_title'],
                'date_joined': entry.get('joined_date') if movement_type == 'arrival' else entry.get('departed_date'),
                'movement_type': movement_type
            })
    return movements
```

```python
engineering_movements = []
for idx, row in df.iterrows():
    engineering_movements += extract_movements(row['recent_exec_hires_parsed'], 'arrival', row['website'])
    engineering_movements += extract_movements(row['recent_exec_departures_parsed'], 'departure', row['website'])
```

```python
engineering_movements_df = pd.DataFrame(engineering_movements)
print(engineering_movements_df)
```

```python
import pandas as pd
import numpy as np
from pandas.tseries.offsets import DateOffset

# Parse sales data and calculate absolute and relative deltas
def extract_sales_data(row):
    sales_data = row['employee_count_by_month_by_role']
    sales_data = eval(sales_data)
    sales_team_data = {k: v.get('sales', 0) for k, v in sales_data.items()}  # Extract sales data
    return sales_team_data

def calculate_sales_deltas(sales_data):
    sales_df = pd.DataFrame(list(sales_data.items()), columns=['month', 'sales'])
    sales_df['month'] = pd.to_datetime(sales_df['month'], format='%Y-%m')
    sales_df = sales_df.set_index('month').sort_index()

    most_recent_date = sales_df.index[-1]
    six_months_ago = most_recent_date - DateOffset(months=6)
    recent_sales = sales_df[sales_df.index > six_months_ago]

    if not recent_sales.empty:
        start_sales = recent_sales.iloc[0]['sales']
        end_sales = recent_sales.iloc[-1]['sales']

        # Calculate absolute and relative delta
        absolute_delta = end_sales - start_sales
        relative_delta = (absolute_delta / start_sales) if start_sales != 0 else float('nan')

        return start_sales, end_sales, absolute_delta, relative_delta
    return None, None, None, None

def normalize_with_percentiles(value, lower_percentile, upper_percentile):
    """Normalize a value to a scale of 0 to 10 based on lower and upper percentiles."""
    if value < lower_percentile:
        return 0
    elif value > upper_percentile:
        return 10
    else:
        return 10 * (value - lower_percentile) / (upper_percentile - lower_percentile)

# Health score calculation
def calculate_health_score(row, lower_percentile_size, upper_percentile_size, lower_percentile_delta, upper_percentile_delta):
    _, end_sales, absolute_delta, relative_delta = calculate_sales_deltas(row['sales_data'])
    if end_sales is not None:
        norm_relative_delta = normalize_with_percentiles(relative_delta, lower_percentile_delta, upper_percentile_delta)
        norm_size = normalize_with_percentiles(end_sales, lower_percentile_size, upper_percentile_size)

        # Weighted average score
        score = 0.6 * norm_relative_delta + 0.4 * norm_size
        return score
    return 0

# Load the DataFrame
df = pd.read_csv('/content/sample_data/pdl_file.csv')

df['sales_data'] = df.apply(extract_sales_data, axis=1)

# Calculate sales values and their statistics for normalization
sales_values = df['sales_data'].apply(lambda x: max(x.values())).dropna()
lower_percentile_size = sales_values.quantile(0.1)
upper_percentile_size = sales_values.quantile(0.9)

# Calculate relative deltas for normalization
relative_deltas = df.apply(lambda row: calculate_sales_deltas(row['sales_data'])[3], axis=1)
lower_percentile_delta = relative_deltas.quantile(0.1)
upper_percentile_delta = relative_deltas.quantile(0.9)

# Apply the health score calculation
df['Health Score'] = df.apply(lambda row: calculate_health_score(row, lower_percentile_size, upper_percentile_size, lower_percentile_delta, upper_percentile_delta), axis=1)

# Find the company with the highest health score
max_score_row = df.loc[df['Health Score'].idxmax()]
print(f"Company with highest health score: {max_score_row['name']} with a score of {max_score_row['Health Score']}")

print(df[['name', 'Health Score']])
```
